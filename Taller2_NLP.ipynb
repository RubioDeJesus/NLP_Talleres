## Hacer Web Scraping de 10 biografías en Wikipedia (en búcle )

```python
import requests
from bs4 import BeautifulSoup

Autores = ['https://es.wikipedia.org/wiki/Jorge_Isaacs','https://es.wikipedia.org/wiki/Antonio_Caballero_Holgu%C3%ADn','https://es.wikipedia.org/wiki/Andr%C3%A9s_Caicedo','https://es.wikipedia.org/wiki/Miguel_de_Cervantes','https://es.wikipedia.org/wiki/Jorge_Luis_Borges','https://es.wikipedia.org/wiki/H%C3%A9ctor_Abad_Faciolince','https://es.wikipedia.org/wiki/Juan_Gossa%C3%ADn','https://es.wikipedia.org/wiki/Piedad_Bonnett','https://es.wikipedia.org/wiki/Rafael_Pombo','https://es.wikipedia.org/wiki/Emma_Reyes_(artista)'] 
for autor in Autores:
    url = autor
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    print("==========================================")
    print(soup.find(id='firstHeading').text)
    print("==========================================")
    # Obtener el primer parrafo 
    text1 = soup.find_all('p')[1]
    # Obtener links link
    refc = text1.find_all('a', href=True, text=True)
    for parrafo in refc:
        print("-------------------------")
        print(parrafo)
        print("Contenido:")
        print(parrafo.contents)
        print("Título:")
        print(parrafo.text)

```
