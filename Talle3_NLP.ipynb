# Punto 1

 ### Quitar palabras vacias

```python
import nltk
#nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
#Limpiar Texto, pasar a minuscula y dividir en lista
token = SoloPalabra.split() # Se requiere en una sola lista
novacias = [palabra for palabra in token if palabra not in stopwords_sp]
novacias
```

### ¿Cuáles son las 10 palabras no vacias más usadas?

```python
import collections

c = collections.Counter(novacias)
print('Top 10:')
for word, count in c.most_common(10):
    print('%s: %7d' % (word, count))
```

### Punto 3:

### Stemming del documento

```python
from nltk.stem.snowball import SnowballStemmer
englishStemmer=SnowballStemmer("english")
spanishStemmer=SnowballStemmer("spanish")
```

```python
texto_stem = []

for palabra in c:
    raiz = spanishStemmer.stem(palabra)
    texto_stem.append(raiz)
texto_stem
```

### ¿Cuáles son las 10 raíces más usadas?

```python
#Realizo el conteo de la lista con las raices 
c = collections.Counter(texto_stem)
print('Top 10:')
for word, count in c.most_common(10):
    print('%s: %7d' % (word, count))
```
